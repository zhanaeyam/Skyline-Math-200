```python
import researchpy as rp
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df
```

```python
df['H4MA1'].value_counts()
df['Mistreatment_From_Adults'] = df['H4MA1']
df['Mistreatment_From_Adults'].value_counts()
df['Mistreatment_From_Adults'] = df['Mistreatment_From_Adults'].loc[ df['Mistreatment_From_Adults'] !='96'] 
df['Mistreatment_From_Adults'] = df['Mistreatment_From_Adults'].loc[ df['Mistreatment_From_Adults'] !='98'] 
df['Mistreatment_From_Adults'] = df['Mistreatment_From_Adults'].loc[ df['Mistreatment_From_Adults'] !=' '] 
df['Mistreatment_From_Adults'].value_counts()
```

```python
df['Mistreatment_From_Adults'].replace( ['1','2','3','4','5','6'], ['one time', 'two time','three to five times','six to ten times','more than ten times','this has never happened'], inplace = True)
df['Mistreatment_From_Adults'].value_counts() 
```

```python
rp.summary_cat(df['Mistreatment_From_Adults'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Mistreatment_From_Adults'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
plt.title("Before your 18th birthday, how often did a parent or other adult caregiver say things that really hurt your feelings or made you feel like you were not wanted or loved?")
df['Mistreatment_From_Adults'].value_counts(normalize = True).plot(kind ='bar', color = 'pink', width = 0.2)
```

```python
df['H4PE4'].value_counts()
df['Frequent_Mood_Swings'] = df['H4PE4']
df['Frequent_Mood_Swings'].value_counts()
df['Frequent_Mood_Swings'] = df['Frequent_Mood_Swings'].loc[ df['Frequent_Mood_Swings'] !='6'] 
df['Frequent_Mood_Swings'] = df['Frequent_Mood_Swings'].loc[ df['Frequent_Mood_Swings'] !='8'] 
df['Frequent_Mood_Swings'] = df['Frequent_Mood_Swings'].loc[ df['Frequent_Mood_Swings'] !=' '] 
df['Frequent_Mood_Swings'].value_counts()
```

```python
df['Frequent_Mood_Swings'].replace( ['1','2','3','4','5'], ['strongly agree','agree','neither agree nor disagree','disagree','strongly disagree'], inplace = True)
df['Frequent_Mood_Swings'].value_counts()
```

```python
rp.summary_cat(df['Frequent_Mood_Swings'])
```

```python
plt.xlabel("Frequent_Mood_Swings")
plt.ylabel("Percent")
plt.title("I have frequent mood swings")
df['Frequent_Mood_Swings'].value_counts(normalize = True).plot(kind ='barh', color = 'red', width = 0.2, alpha = 0.5)
```

```python
df['Frequent_Mood_Swings'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
from scipy.stats import chi2_contingency
```

```python
table = pd.crosstab( df['Mistreatment_From_Adults'], df['Frequent_Mood_Swings'])
```

```python
chiVal, pVal, df, exp = chi2_contingency(table)
print("Chi-squared Value:" , chiVal)
print("P Value: " , pVal )
print("Degress or freedom : " , df)
print("Expected counts :" ,exp.round(1))
```
